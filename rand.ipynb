```python
import pandas as pd
import numpy as np
np.random.seed(251844)
data = pd.read_csv('RegistrationData.csv', index_col=0)
def finder(surn):
    for i in range(5):
        if surn in category1[i]:
            return i
teamlead = data[data["First Choice"] == "Team Leader"]
category1 = [[],[],[],[],[]]
for sur, team in teamlead.iterrows():
    category1[0].append(sur)
creative = data[data["First Choice"] == "Creative Thinker"]
for sur, team in creative.iterrows():
    category1[1].append(sur)
researcher = data[data["First Choice"] == "Researcher"]
for sur, team in researcher.iterrows():
    category1[2].append(sur)
designer = data[data["First Choice"] == "Designer"]
for sur, team in designer.iterrows():
    category1[3].append(sur)
pitcher = data[data["First Choice"] == "Presenter/Pitcher"]
for sur, team in pitcher.iterrows():
    category1[4].append(sur)
category2 = {}
secondchoice = data[["Second Choice"]]
for surname, cat in secondchoice.iterrows():
    category2[str(surname)] = str(cat["Second Choice"])

groupings = [[], [], [], [], [], [], [], [], [], [], [], [], [], [], [], [], [], [], [], [], []]
roles = ["Team Leader", "Creative Thinker", "Researcher", "Designer", "Presenter/Pitcher"]
for k in range(21):
    for i in range(5):
        if category1[i]:
            groupings[k].append(np.random.choice(category1[i]))
            category1[i].remove(groupings[k][-1])
            category2.pop(groupings[k][-1])
        else:
            keys = [key for key, value in category2.items() if value == roles[i]]
            if keys:
                m = np.random.choice(keys, 1, replace=False)
                groupings[k].append(m[0])
                category1[finder(m[0])].remove(m[0])
                category2.pop(groupings[k][-1])
            else:
                i = 4 if i == 3 else(3 if i == 2 else (2 if i == 1 else (1 if i == 0 else 0)))
                keys = [key for key, value in category2.items() if value == roles[i]]
                if keys:
                    m = np.random.choice(keys, 1, replace=False)
                    groupings[k].append(m[0])
                    category1[finder(m[0])].remove(m[0])
                    category2.pop(groupings[k][-1])
                else:
                    i = 4 if i == 3 else(3 if i == 2 else (2 if i == 1 else (1 if i == 0 else 0)))
                    keys = [key for key, value in category2.items() if value == roles[i]]
                    if keys:
                        m = np.random.choice(keys, 1, replace=False)
                        groupings[k].append(m[0])
                        category1[finder(m[0])].remove(m[0])
                        category2.pop(groupings[k][-1])
                    else:
                        i = 4 if i == 3 else(3 if i == 2 else (2 if i == 1 else (1 if i == 0 else 0)))
                        keys = [key for key, value in category2.items() if value == roles[i]]
                        if keys:
                            m = np.random.choice(keys, 1, replace=False)
                            groupings[k].append(m[0])
                            category1[finder(m[0])].remove(m[0])
                            category2.pop(groupings[k][-1])
                        else:
                            i = 4 if i == 3 else(3 if i == 2 else (2 if i == 1 else (1 if i == 0 else 0)))
                            keys = [key for key, value in category2.items() if value == roles[i]]
                            if keys:
                                m = np.random.choice(keys, 1, replace=False)
                                groupings[k].append(m[0])
                                category1[finder(m[0])].remove(m[0])
                                category2.pop(groupings[k][-1])
    
    print(f"Group {k+1} ")
    for g in groupings[k]:
        print(g)
    print("\n")                



        
        

           
```
